```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Python\us_customer_data 1 1.csv") # load the data
```

```python
df.head()
```

```python
df_unique = df.drop_duplicates(subset="customer_id", keep="first") #Removing Duplicates Using Customer ID
```

```python
df.isna().sum()
```

```python
df["name"] = df["name"].astype(str).str.strip().str.title() #Clean Customer Names, Remove extra spaces and Convert to Proper Case
```

```python
df["email"].isna().sum() # it filters null values in the email (specific) column 
```

```python
df["email"] = df.apply(
    lambda row: f"{row['name'].lower().replace(' ', '')}@gmail.com"
    if pd.isna(row["email"]) else row["email"].lower(),
    axis=1
)
```

```python
df.head(28)
```

```python
df["phone"] = df["phone"].astype(str).str.strip().str.title() ##Clean email, Remove extra spaces and Convert to Proper Case
```

```python
df["phone"] = df["phone"].str.replace(r"[A-Za-z]", "", regex="true") ## removing letteres in the rows in phone column
```

```python
df["phone"] = df.apply (
    lambda row:"000-000-0000" if pd.isnull(row["phone"]) else(row["phone"]), axis=1) # adding phone numbers in place of null values
```

```python
df["phone"].isna().sum()
```

```python
df["state_code"] = df["address"].str.extract(r"\b([A-Z]{2})\b") # extract state_code from the address column
```

```python
df.info()
```

```python
df["country_code"] = "US" # creating a new column 'country_code' for mapping
```

```python
df.info()
```

```python
mapping = pd.read_csv(r"C:\Users\91833\Downloads\Country_2-3_Digits_Codes.csv") # loading new dataset for mapping columns
```

```python
mapping.head()
```

```python

country_to_dial = dict(zip(mapping["A2"], mapping["DIALINGCODE"])) # adding dictionary
```

```python
country_to_dial # verify dictionary
```

```python
df["dial_code"] = " + " + df["country_code"].map(country_to_dial) + "-" #mapping country_code, dail_code and creating new column as dail_code
```

```python
df.info()
```

```python
df["phone"] = df["phone"].astype(str).str.replace(r"\D", "", regex=True)
```

```python
df["phone_no"] = df.apply(
    lambda row: f"{row['dial_code']} ({row['phone'][:3]}) {row['phone'][3:6]}-{row['phone'][6:10]}"
    if len(row["phone"]) >= 10 else "INVALID NUMBER",
    axis=1
)
```

```python
df.head()
```

```python
df = df.drop(columns=['phone']) # remove the column phone
```
